```python
#identify the outliers 1
import pandas as pd 
import matplotlib.pyplot as plt 

dataset = pd.read_csv('Flavors.csv')

dataset[['Flavor Rating']].hist(figsize=(3, 4), bins=7, xlabelsize=6, ylabelsize=6)
dataset[['Texture Rating']].hist(figsize=(3, 4), bins=7, xlabelsize=6, ylabelsize=6)
dataset[['Total Rating']].hist(figsize=(3, 4), bins=7, xlabelsize=6, ylabelsize=6)
plt.show()
```

```python
#how to detect outliers using boxplot 1 
import pandas as pd 
import matplotlib.pyplot as plt 

dataset = pd.read_csv('Flavors.csv')

dataset[['Flavor Rating']].plot(kind='box')
plt.title('Box Plot of Flavor Rating', size=10)
plt.show()
```
